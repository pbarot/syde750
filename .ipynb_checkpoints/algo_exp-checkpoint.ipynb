```python

def normalize(signal):
    return  (signal-min(signal))/(max(signal)-min(signal))
```

```python
# plt.plot(data['h'])
# plt.plot(data['w'])
plt.plot(data['h']/data['w'])
```

```python
plt.figure(figsize = (20,10))
plt.plot(normalize(data['num']))
#plt.plot(normalize(data['h']))
plt.plot(normalize(data['g']))
```
